# Set up and WorkFlow

## Importing the Packages

```python
import requests
from bs4 import BeautifulSoup
```

# Making a Get Request

```python
base_site="https://en.wikipedia.org/wiki/Music"
response=requests.get(base_site)
response.status_code
```

```python
html=response.content
html[:500]
```

## Choose the Parser and create the BeautifulSoup Object

```python
soup=BeautifulSoup(html,"html.parser")
```

## Exporting the Html to a file

```python
with open('Wiki_response.html','wb')as file:
    file.write(soup.prettify('utf-8'))
```

# Searching and Navigate the html tree

### Two most Popular ways to searching in BeautifulSoup are find() and find_all()
### The simplest way to search for an element by its Tag Name

```python
# To get the head element

soup.find('head')
```

```python
# if we search for tag which doesnt exist then nothing happens
soup.find('video')
```

```python
links=soup.find_all('a')
```
